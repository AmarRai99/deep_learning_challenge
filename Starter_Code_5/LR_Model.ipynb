```python
# Dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
application_df = pd.read_csv("charity_data.csv")
application_df
```

```python
application_df = application_df.drop(columns=['EIN', 'NAME', 'ORGANIZATION','SPECIAL_CONSIDERATIONS'])

application_df.head(5)
```

```python
numeric_df = pd.get_dummies(application_df)
numeric_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
X = numeric_df.drop("IS_SUCCESSFUL", axis=1)
y = numeric_df["IS_SUCCESSFUL"]

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler().fit(X_train)


# Scale the data
X_train_scaled = scaler.transform(X_train)

X_test_scaled = scaler.transform(X_test)
```

LR Model

```python
clf = LogisticRegression()

clf.fit(X_train_scaled, y_train)
```

```python
# Final Model
print("Logistic Regression Model")

print(f"The Training Data Score: {clf.score(X_train_scaled, y_train)}")

print(f"The Testing Data Score: {clf.score(X_test_scaled, y_test)}")
```
